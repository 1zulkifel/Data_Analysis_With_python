```python
speed_check=pd.read_csv('files/speed_check.csv',delimiter=';',parse_dates=True,index_col='Date')
speed_check
```

```python
speed_check.dtypes
```

```python
speed_check.groupby('Car').mean()
```

```python
monthly=speed_check.groupby(pd.Grouper(freq='M')).mean()
monthly
```

```python
monthly.index=monthly.index.month_name()
```

```python
monthly
```

```python
speed_check.to_excel('files/spdchck.xlsx')
```

```python
#export to excel
data=pd.ExcelWriter("files/speedcheck.xlsx")
speed_check.groupby('Car').mean().to_excel(data,sheet_name='cars')
speed_check.to_excel(data,sheet_name='monthly')
data.close()
```

```python
#Creating live  Chart
data=pd.ExcelWriter("files/liveChart.xlsx")
speed_check.groupby('Car').mean().to_excel(data,sheet_name='Cars')
speed_check.to_excel(data,sheet_name='monthly')
chart1=data.book.add_chart({'type':'column'})
chart1.add_series({
    'name': 'Cars',
    'categories': '=Cars!A$2:$A$4',
    'values':'=Cars!B$2:$B$4',})

data.sheets['Cars'].insert_chart('D2',chart1)

data.close()
```

```python
data=pd.read_csv('files/salesdata.csv',parse_dates=True) #,index_col='Ship Date')
data.head()
```

```python
data=data.drop(['Ship Date','Region','Sales Channel','Order Priority',
                 'Order Date','Order ID','Total Revenue','Total Cost','Total Profit'],axis='columns')
data
```

```python
Date=pd.date_range('1/1/2021',periods=1000)
dates=pd.DataFrame(Date)
data['Dates']=Date
data['Dates'] = pd.to_datetime(data['Dates'])
df=data.set_index('Dates')
df
```

```python
df.groupby('Country').sum()
```

```python
df.dtypes
```

```python
rep_sales=df.groupby('Country').sum()['Units Sold']
rep_sales.head()
```

```python
df.groupby(pd.Grouper(freq='M')).sum()['Units Sold'].head()
```

```python
monthly_sale=df.groupby(pd.Grouper(freq='M')).sum()['Units Sold']
monthly_sale.index=monthly_sale.index.month_name()
monthly_sale.head()
```

```python
workbook=pd.ExcelWriter("SalesReport.xlsx")
rep_sales.to_excel(workbook,sheet_name='Sale Per rep')
monthly_sale.to_excel(workbook,sheet_name='Monthly')

workbook.close()
```

```python
workbook=pd.ExcelWriter("S_Report.xlsx")
rep_sales.to_excel(workbook,sheet_name='Sale Per rep')
monthly_sale.to_excel(workbook,sheet_name='Monthly')

chart1=workbook.book.add_chart({'type':'column'})
chart1.add_series({
    'name':'Sale Per rep',
    'catergories':'=\'Sale Per rep\'!$A$2:$A$186',
    'values':'=\Sale Per rep\'!$B$2:$B$186',
})
workbook.sheets['Sale Per rep'].insert_chart('D2',chart1)

chart1=workbook.book.add_chart({'type':'column'})
chart1.add_series({
    'name':'Monthly Sale',
    'catergories':'Monthly !$A$2:$A$34',
    'values':'Monthly!$B$2:$B$34',
})
workbook.sheets['Monthly'].insert_chart('D2',chart1)

workbook.close()
```
